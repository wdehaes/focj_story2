Twitter conversation around secondhand

```python
with open('twitter/secondhand_fashion.json', 'r') as f:
    tweets = json.load(f)

with open('twitter/secondhand_clothes.json', 'r') as f:
    new_tweets = json.load(f)
tweets = tweets + new_tweets
```

```python
def get_text(tweets):
    return [re.sub(r"http\S+", "", get_tweet_text(x)) for x in tweets] 

def get_tweet_text(tweet):
    if 'full_text' in tweet:
        return tweet['full_text']
    else:
        return tweet['text']

def print_n_tweets(tweets):
    print("total number of tweets: {}".format(str(len(tweets))))
    
def filter_tweet(tweet):
    if 'full_text' in tweet:
        return len(tweet['full_text']) > 10
    if 'text' in tweet:
        return len(tweet['text']) > 10
    return False
```

```python
print_n_tweets(tweets)
tweets = [tweet for tweet in tweets if filter_tweet(tweet)]
```

```python
for tweet in tweets:
    tweet['content'] = nlp(get_tweet_text(tweet))
```

```python
for tweet in tweets:
    tweet['lemmas'] = []
    for word in tweet['content']:
        if not word.is_stop and not word.is_punct:
            tweet['lemmas'].append(word.lemma_)
```

```python
word_count = {}
for tweet in tweets:
    for lemma in tweet['lemmas']:
        if lemma in word_count:
            word_count[lemma] = word_count[lemma] + 1
        else:
            word_count[lemma] = 1
    
```

```python
word_count
```

```python
import collections
sorted_x = sorted(word_count.items(), key=lambda kv: kv[1])
sorted_word_count = collections.OrderedDict(sorted_x.reverse())

plt.bar(range(len(word_count)), sorted_word_count.values(), align='center')
plt.xticks(range(len(word_count)), list(sorted_word_count.keys()))

plt.show()
```
